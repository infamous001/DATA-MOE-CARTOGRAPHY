```python
import matplotlib.pyplot as plt
import numpy as np

# Flatten the nested lists in the dictionaries
data = [val for dct in expert_dist for val in dct.values()]
flattened_lists = [[item for sublist in inner_lst for item in sublist] for inner_lst in data]

# Count the occurrences of each element
counts = np.array([np.unique(lst, return_counts=True)[1] for lst in flattened_lists]).T

# Plot the distributions
plt.figure(figsize=(10, 6))
for expert_idx in range(3):
    plt.bar(counts[expert_idx], color=f'C{expert_idx}', alpha=0.7, label=f'Expert {expert_idx}')
plt.legend()
plt.title('Distribution Over Experts 0, 1, and 2')
plt.xlabel('Elements')
plt.ylabel('Counts')
plt.show()
```
